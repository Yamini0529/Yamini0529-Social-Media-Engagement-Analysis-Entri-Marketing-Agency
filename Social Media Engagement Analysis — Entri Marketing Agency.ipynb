```python
# inheritance
class SocialPost:
    def __init__(self,likes,comments):
        self.likes = likes        
        self.comments = comments  
    def display(self):
            print(f" likes: {self.likes}, comments: {self.comments}")     

SP=SocialPost(100,'Good')
SP.display()
class Author(SocialPost):
    def __init__(self,likes, comments,author_rating):
        super().__init__(likes,comments)    # super fun inheritance allmethods and  properties from parents
        self.author_rating = author_rating
    def display(self):
        print(f" Rating: {self.author_rating}/5")

x= Author(100,'Good',3.2)    
x.display() 

```

```python

class SocialPost:
    def __init__(self,likes,comments):
        self.likes = likes
        self.comments = comments

    def display(self):
        
        print(f"Likes: {self.likes}, Comments: {self.comments}")

# Subclass with additional attribute
class Author(SocialPost):
    def __init__(self,likes, comments, author_rating):
        super().__init__(likes, comments)
        self.author_rating = author_rating

    def display(self):
        super().display()
        print(f"Author Rating: {self.author_rating}/5")

# Test polymorphism
p1 = SocialPost(500,'Good')
p2 = Author(1000,'Excellent', 4.5)

for post in [p1, p2]:
    print(f" Display Post")
    
    post.display()
```

#  Data Visualization 

```python


import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("C:/Users/rajes/OneDrive/Documents/social_media_dataset.xlsx")
df.head()

daily_freq = df['Date_Posted'].dt.date.value_counts().sort_index()

plt.figure(figsize=(12,5))
plt.plot(daily_freq.index, daily_freq.values,linestyle='dotted', marker='*')
plt.title("Daily Post Frequency")
plt.xlabel("Date")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Stacked bar chart of average engagement by category
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("C:/Users/rajes/OneDrive/Documents/social_media_dataset.xlsx")

df['Engagement'] = df['Likes'] + 2*df['Shares'] + 1.5*df['Comments']
avg_engagement = df.groupby('Category')[['Likes', 'Shares', 'Comments']].mean()

avg_engagement.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Average Engagement by Category")
plt.ylabel("Average Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix heatmap
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("C:/Users/rajes/OneDrive/Documents/social_media_dataset.xlsx")
df['Engagement'] = df['Likes'] + 2*df['Shares'] + 1.5*df['Comments']
avg_engagement = df.groupby('Category')[['Likes', 'Shares', 'Comments']].mean()

avg_engagement.plot(kind='bar', stacked=True, figsize=(10,6))

sns.heatmap(df[['Likes', 'Shares', 'Comments', 'Post_Length', 'Sentiment_Score', 'Engagement']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
